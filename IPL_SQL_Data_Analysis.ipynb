```python
import pandas as pd
import sqlite3
```

```python
ipl_df = pd.read_csv('IPL_Matches.csv')
ipl_df.head()
```

Check Available Columns

```python
ipl_df.columns
```

Create SQLite Table

```python
conn = sqlite3.connect(':memory:')
ipl_df.to_sql('ipl', conn, index=False, if_exists='replace')
```

Total Matches Played

```python
pd.read_sql("SELECT COUNT(id) AS total_matches FROM ipl", conn)
```

Matches Played Each Season

```python
pd.read_sql("""
SELECT season, COUNT(*) AS matches_played
FROM ipl
GROUP BY season
ORDER BY season
""", conn)
```

Most Successful Teams by no.of wins

```python
pd.read_sql("""
SELECT winner, COUNT(*) AS total_wins
FROM ipl
GROUP BY winner
ORDER BY total_wins DESC
""", conn)
```

Most Toss Wins

```python
pd.read_sql("""
SELECT toss_winner, COUNT(*) AS toss_wins
FROM ipl
GROUP BY toss_winner
ORDER BY toss_wins DESC
""", conn)
```

Toss Decision Stats (Bat or Field)

```python
pd.read_sql("""
SELECT toss_decision, COUNT(*) AS count
FROM ipl
GROUP BY toss_decision
""", conn)
```

Top 10 Players of the Match

```python
pd.read_sql("""
SELECT player_of_match, COUNT(*) AS awards
FROM ipl
GROUP BY player_of_match
ORDER BY awards DESC
LIMIT 10
""", conn)
```

Most Popular Venues

```python
pd.read_sql("""
SELECT venue, COUNT(*) AS match_count
FROM ipl
GROUP BY venue
ORDER BY match_count DESC
LIMIT 10
""", conn)
```

Team vs Team Match Count

```python
pd.read_sql("""
SELECT team1, team2, COUNT(*) AS total_encounters
FROM ipl
GROUP BY team1, team2
ORDER BY total_encounters DESC
""", conn)
```

Matches Per Year and Winner

```python
pd.read_sql("""
SELECT season, winner, COUNT(*) AS wins
FROM ipl
GROUP BY season, winner
ORDER BY season, wins DESC
""", conn)
```

Matches with No Result

```python
pd.read_sql("""
SELECT COUNT(*) AS no_result_matches
FROM ipl
WHERE result = 'no result'
""", conn)
```
